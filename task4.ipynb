```python
#133
numbers = {
        "one": "1",
        "two": "2",
        "three": "3",
        "four": "4",
        "five": "5",
        "six": "6",
        "seven": "7",
        "eight": "8",
        "nine": "9",
        "zero": "0"
}
def defineNumber(number):
    for numba in numbers:
        if (number == int(numbers[numba])):
            return numba

def define(number):
    ans = ''
    intNumber = int(number)
    ok = False
    if (100 <= intNumber <= 999):
        ans += defineNumber(int(intNumber/100)) + " hundred "
        intNumber %= 100
        if (intNumber % 100 == 0):
            ok = True
    if (20 <= intNumber < 100):
        if (int(intNumber / 10) == 2):
            ans += "twenty " 
        elif(int(intNumber / 10) == 3):
            ans += "thirty "
        elif(int(intNumber / 10) == 5):
            ans += "fifty "
        else:
            ans += defineNumber(int(intNumber / 10)) + "ty "
        intNumber %= 10
        if (intNumber % 10 == 0):
            ok = True
    if (10 <= intNumber < 20 and intNumber != 0):
        ok = True
        if (intNumber == 10):
            ans += "ten "
        elif (intNumber % 10 == 1):
            ans += "eleven "
        elif (intNumber % 10 == 2):
            ans += "twelve "
        elif (intNumber % 10 == 3):
            ans += "thirteen "
        elif (intNumber % 10 == 5):
            ans += "fifteen "
        else:
            ans += defineNumber(intNumber%10) + "teen "
        intNumber %= 10
    if (intNumber < 10 and not ok):
        ans += defineNumber(intNumber)
        intNumber /= 10
    return ans

if "__main__":
    n = input()
    ans = define(n)
    res = ''
    for i in ans:
        for number in numbers:
            if i == numbers[number]:
                res += number
        res += i

print(res)
```
